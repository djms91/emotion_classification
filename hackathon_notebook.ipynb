## Methodology

The text data was preprocessed by cleaning and normalizing the labels. The sentences were then converted into numerical form using TF-IDF (Term Frequency–Inverse Document Frequency) vectorization.

A Logistic Regression model was used for classification. To handle class imbalance in the dataset, the parameter( class_weight = balanced )was applied.

The dataset was split into training and testing sets using an 80-20 ratio to evaluate model performance

```python
import pandas as pd
y_train=pd.Series(y_train,name="emotion")
x_train=pd.Series(x_train,name="text")
x_train=x_train.reset_index(drop=True)
y_train=y_train.reset_index(drop=True)
train_df=pd.concat([x_train,y_train],axis=1)
```

```python

from sklearn.utils import resample
max_size=train_df["emotion"].value_counts().max()
df_list=[]
for emotion in train_df["emotion"].unique():
    df_class=train_df[train_df["emotion"]==emotion]
    df_upsampled=resample(
        df_class,
        replace=True,
        n_samples=max_size,
        random_state=42)
    df_list.append(df_upsampled)
df_balanced=pd.concat(df_list)
x_train_balanced=df_balanced["text"]
y_train_balanced=df_balanced["emotion"]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

clf = Pipeline([
    ("tfidf", TfidfVectorizer(stop_words='english', ngram_range=(1,2))),
    ("model", LogisticRegression(max_iter=1000,class_weight="balanced"))
])
```

## Results

The model achieved an accuracy of around 60% on the test data. It performs best on negative sentiment, 
while neutral and positive classes show moderate performance.
The model shows difficulty in understanding implicit and context-based expressions, which is a common challenge in natural language processing tasks.


```python
y_pred=clf.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
